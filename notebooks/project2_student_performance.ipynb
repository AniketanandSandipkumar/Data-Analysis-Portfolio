1. Key Findings
1. Previous grades strongly determine final performance  
2. Higher absences reduce performance  
3. Study time positively impacts scores  
4. Pass rate is significantly influenced by attendance  

2. Recommendations
1. Early intervention for low G1/G2 students  
2. Attendance monitoring system
